```python
from astropy.io import fits
from astropy.table import Table
from astropy.wcs import WCS

from astropy.nddata import Cutout2D

from astropy.visualization import ZScaleInterval

import numpy as np

from numpy.random import randint

from matplotlib import pyplot as plt

import GalPRIME.galprime as galprime
from GalPRIME.galprime import plotting as plotter

from IPython.display import clear_output
```

```python
t = Table.read("COSMOS_V9_CLEAN.fits")
```

```python
def select_backgrounds(file_directory, object_table, n_bgs, img_size = 251, bright=25,
                      ra_key="RA_1", dec_key="DEC_1", mag_key="i", arcconv=0.168):
    backgrounds, bg_info = [], []
    
    img_list = galprime.get_image_filenames(file_directory)

    
    def cutout_select(data, wcs, size):
        rand_x = randint(size / 2 + 1, data.shape[0] - (size / 2) - 1)
        rand_y = randint(size / 2 + 1, data.shape[1] - (size / 2) - 1)

        cutout = Cutout2D(data, (rand_x, rand_y), size, wcs=wcs)

        return cutout
    
    def cut_table(t, ras, decs, ra_key="RA", dec_key="DEC"):
        ras, decs = np.sort(ras), np.sort(decs)
        t_cut = t[t[ra_key] > ras[0]]
        t_cut = t_cut[t_cut[ra_key] < ras[1]]
        t_cut = t_cut[t_cut[dec_key] > decs[0]]
        t_cut = t_cut[t_cut[dec_key] < decs[1]]

        return t_cut
        
    # Here is when we go through the potential backgrounds
    while len(backgrounds) < n_bgs:
        clear_output()

        print(len(backgrounds), "backgrounds selected so far")
        # Select random image
        img_fn = np.random.choice(img_list)
        with fits.open(img_fn) as HDUList:
            w = WCS(HDUList[1].header)
            data = HDUList[1].data
        
        # Load the cutout, and trim the table to get the objects within the frame
        cutout = cutout_select(data, w, img_size)
        cutout_wcs = cutout.wcs
        ra_min, dec_min = cutout_wcs.wcs_pix2world(0, 0, 0)
        ra_max, dec_max = cutout_wcs.wcs_pix2world(cutout.data.shape[0], cutout.data.shape[1], 0)
        t_cut = cut_table(t, (ra_min, ra_max), (dec_min, dec_max), 
                          ra_key="RA_1", dec_key="DEC_1")
        t_bright = t_cut[t_cut[mag_key] < bright]
        
        positions = cutout_wcs.wcs_world2pix(t_cut[ra_key], t_cut[dec_key], 0)
        positions_bright = cutout_wcs.wcs_world2pix(t_bright[ra_key], t_bright[dec_key], 0)

        center = (cutout.data.shape[0]/2, cutout.data.shape[1]/2)
        max_dist = np.sqrt(center[0] ** 2 + center[1] ** 2)
        distances = np.sqrt((center[0] - positions[0]) ** 2 + (center[1] - positions[1]) ** 2)
        distances_bright = np.sqrt((center[0] - positions_bright[0]) ** 2 + (center[1] - positions_bright[1]) ** 2)
        
        
        plt.figure(figsize=(5,5))
        lims = ZScaleInterval(contrast=0.5).get_limits(cutout.data)
        plt.imshow(cutout.data, vmin=lims[0], vmax=lims[1], cmap="Greys", origin="lower")
        plt.scatter(positions[0], positions[1], color="red", s=2, alpha=0.3)
        plt.scatter(positions_bright[0], positions_bright[1], color="red", s=10)
        
        plt.scatter(center[0], center[1], color="blue", marker="x", alpha=0.5)
        
        plt.tight_layout()
        plt.show()
        
        print("Number of objects in frame:", len(t_cut))
        print("Number of objects with m <", str(bright), "in frame:", len(t_bright)) 
        print("Distance to nearest object:", np.min(distances * arcconv))
        print("Distance to nearest bright object:", np.min(distances_bright * arcconv))

        good_bg = int(input("Accept? :"))
        if good_bg == 1:
            backgrounds.append(cutout.data)
        elif good_bg == 9:
            break
        else:
            continue
    print("Finished getting backgrounds!")
    return backgrounds
        
bgs = select_backgrounds("images/", t, 10, img_size=451)
```
